```python
import numpy as np

def randomTimeSeries(nr, nc, min, max):
  d = np.random.uniform(min, max,size=(nr, nc))
  ret = np.zeros((nr, nc))
  for j in range(0, nc, 1):
    ret[0][j] = d[0][j]
    for i in range(1, nr, 1):
      ret[i][j] = ret[i-1][j]+d[i][j];
          
  return ret;
```

# **by Using Simulated annealing to mimimize the cost**

```python
from math import exp
import random
def simulated_annealing (Temperature, numberOfIterations, coolingRate, length):
  bestCost = cost(cm[:,0], cm[:,1])
  bestSignal = cm[:,1]
  for i in range(numberOfIterations):
    if Temperature > 0.1:

      right = np.roll(bestSignal, 1)
      left = np.roll(bestSignal, -1)
      if bestSignal.any() + 0.1 > 100: up = bestSignal 
      else: up = bestSignal + 0.1
      if bestSignal.any() - 0.1 < 0: down = bestSignal 
      else: down = bestSignal - 0.1

      currentSignal = random.choice([right, left, up, down])
      currentCost = cost(cm[:,0], currentSignal)
      if currentCost < bestCost :
        bestSignal = currentSignal
        bestCost = currentCost
      elif exp((bestCost - currentCost)/Temperature) < random.random() :
        pass
      Temperature *= coolingRate  
    #if Temperature % 100 == 0:  print('temp: {} numberOfIterations: {}  cost: {}'.format(Temperature, numberOfIterations, bestCost)) 
      
  return bestSignal  
```
